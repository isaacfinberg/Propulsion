```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

# 1a

```python
gamma = 1.4
R = 8.31
f = 0.
T04 = 2500
z = 10e3
Ts = 300.
Ps = 101e3
rhos = 1.18
g = 9.81
```

```python
Ta = Ts * ( 1 - ( ( gamma - 1 ) / gamma ) * rhos * g * z / Ps )
Pa = Ps * ( Ta / Ts )**( gamma / ( gamma - 1 ) )
M = numpy.linspace( 0.01, 6, 100 )
```

```python
T_mdot = ( M * numpy.sqrt( gamma * R * Ta ) * ( ( ( 1 + f ) * numpy.sqrt( T04 / Ta ) 
           * ( 1 + ( ( gamma - 1 ) / gamma ) * M**2 )**( -1 / 2 ) - 1 ) ) )
T_mdot_Mmax = M[ numpy.where( T_mdot == max( T_mdot ) )[ 0 ][ 0 ] ]
T_mdot_max = max( T_mdot )
print( T_mdot_max, T_mdot_Mmax )
```

```python
pyplot.plot( M, T_mdot )
pyplot.grid()
pyplot.xlabel( 'Mach Number' )
pyplot.ylabel( 'Thrust per Mass Flow Rate (Ns/kg)' );
```

# 1b

```python
u = M * numpy.sqrt( gamma * R * Ta )
eta_p = 1 / ( 1 + T_mdot / ( 2 * u ) )
```

```python
pyplot.plot( M, eta_p )
pyplot.grid()
pyplot.xlabel( 'Mach Number' )
pyplot.ylabel( 'Propulsive Efficiency' );
```
