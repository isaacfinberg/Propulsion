```python
import numpy
from matplotlib import pyplot
%matplotlib inline
```

# 1b

```python
Ts = 300
Ps = 101000
rhos = 1.18 
g = 9.81
gamma = 1.4
zm = numpy.linspace( 0, 12000, 1000 )
zft = numpy.linspace( 0, 39370.1, 1000 )
T = ( 9 / 5 * Ts - 459.67 ) * ( 1 - ( gamma - 1 ) / gamma * ( .062428 * rhos * 3.28084 * g * zft ) / ( 0.0208854 * Ps ) )
rho = rhos * ( 1 - ( gamma - 1 ) / gamma * ( rhos * g * zm ) / Ps )**( 1 / ( gamma - 1 ) )
P = Ps * ( 1 - ( gamma - 1 ) / gamma * ( rhos * g * zm ) / Ps )**( gamma / ( gamma - 1 ) )
```

```python
pyplot.plot( zm, P * .01 )
pyplot.grid()
pyplot.ylabel( 'Pressure (mbar)' )
pyplot.xlabel( 'Altitude (m)' );
```

<font size="5">This model matches with the balloon measurements and US Standard Atmosphere very well.</font>

```python
pyplot.plot( zft, T )
pyplot.grid()
pyplot.ylabel( 'Temperature (F)' )
pyplot.xlabel( 'Altitude (ft)' );
```

<font size="5">This model does not match very well with the balloon measurements and US Standard Atmosphere because the magnitude of the slope of the curve is much larger.</font>

```python
pyplot.plot( zm, rho )
pyplot.grid()
pyplot.xlabel( 'Altitude (m)' )
pyplot.ylabel( 'Density (kg/m^3)' );
```

# 2 Part 2

```python
g = 9.81
eta0 = .4
Qr = 120900e3
LD = numpy.linspace( 15, 20, 6 )
S = numpy.linspace( 0, 20000e3, 1000 )
m1m2 = numpy.zeros( ( len( LD ), len( S ) ) )
for i in range( len( LD ) ):
    m1m2[ i, : ] = numpy.exp( ( g * S ) / ( eta0 * LD[ i ] * Qr ) )
```

```python
label = numpy.array( [ 'L/D = 15', 'L/D = 16', 'L/D = 17', 'L/D = 18', 'L/D = 19', 'L/D = 20' ] )
for i in range( len( LD ) ):
    pyplot.plot( S / 1000, m1m2[ i, : ], label = label[ i ] )
pyplot.grid()
pyplot.legend()
pyplot.ylabel( 'm1/m2' )
pyplot.xlabel( 'Range (km)' );
```
