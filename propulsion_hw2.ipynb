```python
import numpy
from matplotlib import pyplot 
%matplotlib inline
```

```python
gamma = numpy.linspace( 1.2, 1.7, 50 )
P0 = 100 * 101325
R = 8.31
T0 = 1000
mdot_A = ( P0 / numpy.sqrt( R * T0 ) * numpy.sqrt( gamma ) * 
         ( 2 / ( gamma + 1 ) )**( ( gamma + 1 ) / ( 2 * gamma - 2 ) ) )
```

```python
pyplot.plot( gamma, mdot_A )
pyplot.grid()
pyplot.xlabel( 'Specific Heat Ratio' )
pyplot.ylabel( ' Normalized Mass Flow Rate (kg/sm^2)');
```

```python
M1 = numpy.linspace( 1, 10., 100 )
M2 = numpy.zeros( ( 3, len( M1 ) ) )
g1 = 1.3
g2 = 1.5
g3 = 1.7
g = numpy.array( [ g1, g2, g3 ] )
```

```python
for i in range( len( g ) ):
    M2[ i, : ] = numpy.sqrt( ( ( g[ i ] - 1 ) * M1**2 + 2 ) / ( 2 * g[ i ] * M1**2 - g[ i ] - 1 ) )
```

```python
label = [ 'gamma = 1.3', 'gamma = 1.5', 'gamma = 1.7' ]
for i in range( len( g ) ):
    pyplot.plot( M1, M2[ i, : ], label = label[ i ] )
pyplot.grid()
pyplot.legend()
pyplot.xlabel( 'M1' )
pyplot.ylabel( 'M2' );
```
